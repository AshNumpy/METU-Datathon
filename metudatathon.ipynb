```python
import pandas as pd 

sales = pd.read_excel("Data23.xlsx",sheet_name = 'SALES')
products = pd.read_excel("Data23.xlsx",sheet_name = "PRODUCTS")
customers = pd.read_excel("Data23.xlsx",sheet_name = 'CUSTOMERS')
```

```python
products.head()
```

```python
merged = pd.merge(pd.merge(sales,products, on = 'ProductID'),customers, on = 'UserID')
```

```python
merged.head()
```

```python
data = merged.sort_values(by = 'DateTime',ascending = True)
data
```
